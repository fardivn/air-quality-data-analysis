## Data Wrangling

### Gathering Data

Langkah pertama yang perlu dilakukan adalah membaca seluruh file .csv pada folder `data`, kemudian disimpan pada list `csv_files`.

```python
csv_files = [file for file in os.listdir("./data") if file.endswith(".csv")]
csv_files
```

Selanjutnya, mengimpor files tersebut menjadi DataFrame dengan menggunakan method `read_csv()`, kemudian disimpan tiap DataFrame-nya ke dalam list `df_list`.

```python
df_list = []
for file in csv_files:
    df = pd.read_csv("./data/" + file)
    df_list.append(df)
```

Setelah itu menggabungkan (concatenate) seluruh DataFrame yang ada di list `df_list` menjadi satu DataFrame.

```python
all_data = pd.concat(df_list, ignore_index=True, sort=False)
```

Berikut adalah hasil penggabungan seluruh dataset.

```python
all_data
```

### Assessing Data

Pada langkah ini, tipe data tiap kolom pada dataset akan diperiksa menggunakan method `info()`.

```python
all_data.info()
```

Dari output di atas dapat dilihat bahwa terdapat data yang bernilai `null` yang terlihat dari jumlah `Non-Null Count`-nya yang beragam. Selain itu, tipe data sudah sesuai dengan apa yang diinginkan.

Berikut adalah jumlah data yang bernilai `null` pada setiap kolomnya.

Berikut adalah baris-baris data yang bernilai `null` pada kolom `CO`.

Dapat dilihat bahwa missing value hanya terjadi pada jam-jam tertentu (tidak sepanjang hari nilainya `null`). 

Berikut adalah jumlah baris yang duplikat.

```python
all_data.duplicated().sum()
```

Dapat dilihat bahwa ternyata pada data tidak terdapat baris data yang duplikat.

Berikut adalah ringkasan parameter statistik dari setiap kolom **numerik** yang ada.

### Cleaning Data

**Drop Column**

Pada `all_data`, kolom `No` tidak lagi berguna karena sudah digantikan dengan `index` yang secara otomatis ditambahkan oleh sistem ketika mengimport data source dari .csv ke DataFrame yang selanjutnya digabung menjadi satu. Pada dasarnya, `No` hanya berguna sebagai index di data source yang terpisah-pisah sebelumnya. Maka dari itu, kolom `No` dapat dihapus (drop).

```python
all_data.drop(columns='No', inplace=True)
all_data
```

**Missing Values**

Dari tahap assessing data, dapat disimpulkan bahwa bisa dilakukan mengisian (fill) data yang hilang dengan dengan nilai sebelumnya atau bisa disebut forward fill. Method yang dapat digunakan pada kasus ini adalah `ffill()`.

```python
all_data.ffill(inplace=True)

# mengecek apakah masih ada data yang null setelah dilakukan forward fill
all_data.isna().sum()
```

Dapat dilihat bahwa data sudah tidak ada yang bernilai `null`, tidak ada data yang duplikat, dan kolom yang tidak penting sudah dihapus, maka proses pembersihan data sudah dapat dikatakan cukup.

## Exploratory Data Analysis (EDA)

```python
all_data.describe(include="all")
```

### Explore Data untuk Temperatur

- Mengetahui rata-rata suhu untuk setiap wilayah setiap tahunnya

```python
all_data.groupby(["station", "year"])["TEMP"].mean().round(decimals=2).unstack()
```

Dari data di atas, dapat dilihat bahwa temperatur cenderung turun setiap tahunnya. Namun perlu menjadi catatan bahwa pada tahun `2017` data hanya sampai bulan Februari saja, bukan sepanjang tahun.

- Mengetahui station mana yang cenderung lebih dingin daripada station lainnya

```python
all_data.groupby("station")["TEMP"].mean().sort_values(ascending=True)
```

Dari output di atas, ternyata station yang cenderung paling dingin adalah Huairou yang berada di angka 12-an derajat celcius, sedangkan pada station lainnya, temperaturnya tidak jauh beda yaitu berada di angka 13-an derajat celcius. Keadaan tersebut bisa jadi karena wilayah Huairou berada di dataran yang lebih tinggi daripada station lain.

- Mengetahui temperatur setiap tahunnya

```python
all_data.groupby("year").agg({
    "TEMP": ["min", "max", "mean"]
}).round(2)
```

Dapat dilihat bahwa suhu terendah adalah -19.90&deg;C yang terjadi pada tahun 2016, sedangkah suhu tertinggi adalah 41.6&deg;C yang terjadi pada tahun 2014.

Dari hasil di atas, dapat dijabarkan lagi menjadi per bulan.

- Mengetahui suhu setiap bulannya pada seluruh wilayah?

```python
all_data.groupby(["year", "month"]).agg({
    "TEMP": ["min", "max", "mean"]
}).round(2)
```

Dari hasil di atas bahwa pada awal dan akhir tahun, suhu cenderung lebih rendah daripada biasanya.

### Explore Data untuk Tingkat Polusi

Sebagai pendahuluan, tingkat polusi dapat dilihat dari kolom/atribut berikut yang merupakan polutan udara.
- `PM2.5`: (Fine Particulate Matter) Partikel udara yang berukuran lebih kecil dari 2.5 mikron
- `PM10`: (Coarse Pariculate Matter) Partikel udara yang berukuran lebih kecil dari 10 mikron
- `SO2`: Sulfur Dioksida
- `NO2`: Nitrogen Dioksida
- `CO`: Karbon Monoksida
- `O3`: Ozon

Semakin tinggi tingkat polutan, maka semakin buruk kualitas udaranya.

<small>Sumber: <a href=https://id.wikipedia.org/wiki/Indeks_kualitas_udara>Wikipedia</small>

```python
polutan_temp = ["PM2.5", "PM10", "SO2", "NO2", "CO", "O3", "TEMP"]
```

- Mengetahui rata-rata polutan udara setiap tahunnya.

```python
all_data.groupby(["year", "month"])[polutan_temp].mean().round(decimals=2)
```

- Mengetahui rata-rata polutan udara pada setiap station.

```python
all_data.groupby("station")[polutan_temp].mean().round(decimals=2)
```

- Mengetahui korelasi antara tingkat polutan dengan temperatur.

```python
all_data[polutan_temp].corr().round(4)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana trend rata-rata temperatur setiap bulannya?

```python
trend_temp = all_data.groupby(["year", "month"])["TEMP"].mean().round(2)
ax = trend_temp.plot(marker=".")
plt.title(f"Rata-Rata Temperatur Per Bulan (\N{DEGREE SIGN}C)")
plt.xlabel(xlabel="Tahun, Bulan")
plt.gcf().set_size_inches(13, 5)
ax.set_xticks(range(len(trend_temp)))
ax.set_xticklabels(trend_temp.index, rotation=90)
plt.grid()
plt.show()
```

### Pertanyaan 2: Bagaimana rata-rata temperatur berdasarkan stationnya?

```python
temp_station = all_data.groupby("station").mean(numeric_only=True).round(2)
sns.barplot(
    y="station", 
    x="TEMP",
    data=temp_station.sort_values(by="TEMP")
)
plt.title(f"Rata-Rata Temperatur Berdasarkan Station (\N{DEGREE SIGN}C)")
plt.xlabel(None)
plt.ylabel(None)
plt.gcf().set_size_inches(10, 5)
plt.show()
```

### Pertanyaan 3: Bagaimana trend rata-rata polutan udara setiap bulannya?
